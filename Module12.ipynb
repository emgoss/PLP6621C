**3. Check for recombination.**

The first analysis we will do is GARD to check for recombination in our data. All of the following analyses should be done on non-recombinant sequences. GARD will identify recombination and associated breakpoints to divide the alignment into non-recombinant partitions.

The input required for GARD, and for most of the HyPhy analyses, is the in-frame coding sequence alignment and an associated newick tree. You can run help to check this.

```python
!hyphy gard --help
```

In this case, I inferred the tree using RAxML for you. (If you use DataMonkey, it will generate a tree for you.)

I was encountering a numerical error in the 2-breakpoint analysis and HyPhy told me to deal with this using ENV=TOLERATE_NUMERICAL_ERRORS=1.

```python
!hyphy gard --alignment rice_yellow_mottle_no_stop.fasta --tree rice_yellow_mottle_RAxML.newick ENV=TOLERATE_NUMERICAL_ERRORS=1;
```
